```python
# 初回のみライブラリをインポートする
import pandas as pd

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("/content/sample_pandas_6.csv")

# 先頭から5行目までを表示する
df.head()
```

```python
# 先頭から2番目にある「Z4WOOIYV」という商品番号の行のみ取得
df.query('商品番号 == "Z4WOOIYV"')
```

```python
# 600円の行のみを抽出
df.query('単価 == 600')
```

```python
# 複数条件に合う行を抽出
# 在庫数が5つ以下の行をすべて取得する
df.query('在庫 <= 5')
```

```python
# 複数条件に合う行を抽出
# ある商品の在庫が５以下＝AND演算子を使う
# 「8T7D5DQA」という商品番号の商品の在庫5以下である行を抽出

df.query('商品番号 == "8T7D5DQA" and 在庫 <= 5')
```

```python
# 分析しやすいようにデータを変形しよう
# 発注商品の単価に消費税を掛けて取り出す

def tax(x):
  return x * 1.10
df['単価'].apply(tax)
```

```python
tax_series = df['単価'].apply(tax)
tax_series.name = "単価（税込み）"
pd.concat([df, tax_series], axis=1)
```
